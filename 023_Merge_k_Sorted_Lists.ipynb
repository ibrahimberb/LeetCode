# 23. Merge k Sorted Lists

[https://leetcode.com/problems/merge-k-sorted-lists/description/](https://leetcode.com/problems/merge-k-sorted-lists/description/)

You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

**Example 1:**
```
Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
```

**Explanation:**
```
The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted list:
1->1->2->3->4->4->5->6
```

**Example 2:**
```
Input: lists = []
Output: []
```

**Example 3:**
```
Input: lists = [[]]
Output: []
```

**Constraints:**
```
k == lists.length
0 <= k <= 104
0 <= lists[i].length <= 500
-104 <= lists[i][j] <= 104
lists[i] is sorted in ascending order.
The sum of lists[i].length will not exceed 104.
```

### Solution #1

````python
from typing import List


class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class ListHelper:
    def __init__(self, LN):
        self.LN = LN
        self.index = 0

    def next(self):
        self.LN = self.LN.next
        self.index += 1

    def get_current_val(self):
        if self.LN is None:
            return None
        return self.LN.val

    def is_exhausted(self):
        return self.LN is None

    def __eq__(self, __value):
        return self.get_current_val() == __value.get_current_val()

    def __gt__(self, __value):
        return self.get_current_val() > __value.get_current_val()


def min_list_helpers(itlh):
    itlh_filtered = {}
    for i, lh in itlh.items():
        if lh.get_current_val() is None:
            continue

        itlh_filtered[i] = lh.get_current_val()

    return min(itlh_filtered, key=itlh_filtered.get)


def merge_k_lists(lists):
    index_to_list_helpers = {i: ListHelper(L) for i, L in enumerate(lists)}
    head = None
    curr = None
    while not all(lh.is_exhausted() for lh in index_to_list_helpers.values()):
        # find the LH that has the smallest in it.
        i = min_list_helpers(index_to_list_helpers)
        val = index_to_list_helpers[i].get_current_val()
        new_node = ListNode(val)
        print("val: ", val)
        if head is None:
            head = new_node
            curr = head
        else:
            curr.next = new_node
            curr = new_node

        index_to_list_helpers[i].next()

    return head


ln_1 = ListNode(1)
ln_1.next = ListNode(2)
ln_1.next.next = ListNode(3)

ln_2 = ListNode(1)
ln_2.next = ListNode(4)
ln_2.next.next = ListNode(5)
````

### Solution #2: Using heaps

````python
import heapq


def merge_k_sorted_lists(lists):
    # Define comparison functions for ListNode objects
    ListNode.__eq__ = lambda self, other: self.val == other.val
    ListNode.__lt__ = lambda self, other: self.val < other.val

    # Create a heap to store the smallest nodes from each list
    node_heap = []

    # Create a head node and a tail node to build the merged list
    head = curr = ListNode(0)

    # Add the first node from each list to the heap
    for node in lists:
        if node:
            heapq.heappush(node_heap, (node.val, node))

    # Pop the smallest node from the heap and add it to the merged list
    while node_heap:
        val, node = heapq.heappop(node_heap)
        curr.next = node
        curr = curr.next
        if node.next:
            heapq.heappush(node_heap, (node.next.val, node.next))

    # Return the merged list
    return head.next


ln_1 = ListNode(1)
ln_1.next = ListNode(2)
ln_1.next.next = ListNode(3)

ln_2 = ListNode(1)
ln_2.next = ListNode(4)
ln_2.next.next = ListNode(5)

lists = [ln_1, ln_2]
result = merge_k_sorted_lists(lists)
while result.next:
    print(result.val)
    result = result.next
````
